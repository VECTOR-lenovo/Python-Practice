**Plot the graph for linear system b and write the point of intersection between three lines.**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

y1 = (3 - 2 * x) / 3  # Line 1: 2x + 3y = 3
y2 = (6 - 4 * x) / 6  # Line 2: 4x + 6y = 6
y3 = (3 * x - 1) / 2  # Line 3: 3x - 2y = 1
fig, ax = plt.subplots()

ax.plot(x, y1, color='r', label='$2x + 3y = 3$')
ax.plot(x, y2, color='g', label='$4x + 6y = 6$')
ax.plot(x, y3, color='b', label='$3x - 2y = 1$')
ax.text(-4, 1.5, '$2x + 3y = 3$', color='r')
ax.text(-4, -1, '$4x + 6y = 6$', color='g')
ax.text(2, 2, '$3x - 2y = 1$', color='b')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)
ax.set_xticks(list(range(-5, 6)))
ax.set_yticks(list(range(-5, 6)))
ax.set_aspect('equal')
ax.grid(True, linestyle=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Graph of Linear System")
plt.show()
```

**3.Create a 4x4 matrix containing random numbers in the range of 1 to 20. Have the user prompted for a number, check that the imputed value is a number and then check if the number is in the matrix. If the number is in the matrix, return the number of times that number is in the matrix.**

```python
import numpy as np
matrix = np.random.randint(1, 21, size=(4, 4))
print("Matrix:")
print(matrix)
number = input("Enter a number between 1 and 20: ")
if number.isdigit():
    number=int(number)
    if number < 1 or number > 20:
        print(f"Number out of range.")
else:
    print(f"Invalid input. Please enter a number.")
count = np.count_nonzero(matrix == number)
if count > 0:
    print(f"The number {number} is in the matrix {count} times.")
else:
    print(f"The number {number} is not in the matrix.")
```

**1. Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?**

```python
import numpy as np
A = np.random.randint(1, 21 , size=(5, 5))
rank_A = np.linalg.matrix_rank(A)
print("Matrix A:")
print(A)
print("Rank of A:", rank_A)
det_A=np.linalg.det(A)
if det_A==0:
    print("Matrix A is not invertible.")
else:
    print("Matrix A is invertible.")
```
