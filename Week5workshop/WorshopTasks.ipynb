b. What are the differences between NumPy array and Python list?

**Ans:**

- **Data type storage**:
  - Python list: Can store elements of different data types.
  - NumPy array: Stores elements of the same data type.

- **Importing modules**:
  - Python list: No need to import any module.
  - NumPy array: Requires importing the NumPy module.

- **Numerical operations**:
  - Python list: Requires explicit loops for element-wise operations.
  - NumPy array: Supports vectorized operations, making numerical computations faster and more concise.

- **Memory consumption**:
  - Python list: Consumes more memory due to the overhead of storing different data types.
  - NumPy array: Consumes less memory as it stores elements of the same data type in a contiguous block.

- **Fast and convenient to use**:
  - Python list: Slower for numerical operations due to the need for explicit loops.
  - NumPy array: Faster and more convenient for numerical operations due to vectorization and optimized C-based implementation.

**Example 7**

**Create two vectors in 2-dimensional (2D) space and plot them using Matplotlib.**

**Solution: A vector in 2D space can created as a 1D ndarray with two elements as:**

**V = np.array([1,1])
W = np.array([-2.0,-1])**

To plot the above vectors, we will use the quiver() function from the pyplot module of Matplotlib library. The quiver() function is used to create 2D field or velocity plots and is perfect for our purpose of vector plotting. Here is an example of plotting the above vectors**

```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

**c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of the vector be U = (1.5, 1.5, 2.0).**


```python
import math

U = (1.5, 1.5, 2.0)
magnitude = 0
for i in U:
    magnitude += i**2
magnitude = math.sqrt(magnitude)
print(magnitude)
```

**d. Write a function to calculate the direction cosines of the vector. For example, let the vector be O = (1.5, 1.5, 2.0).**

```python
import math
O=(1.5, 1.5, 2.0)
magnitude=0
for i in O:
    magnitude += i**2
magnitude = math.sqrt(magnitude)
for i in O:
    print(i/magnitude)
```

**e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors be a= (1.5, 1.5, 2.0) and b = (1.5, 0.5, 2.0).**

```python
import math
a = (1.5, 1.5, 2.0)
b = (-1.5, 0.5, 2.0)
dot_product = 0
for i in range(len(a)):
    dot_product += a[i] * b[i]
print(dot_product)
```
